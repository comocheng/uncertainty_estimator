# Scenario 1: ads tree node trained on 1 vs 2 vs 4 species

More data means lower variance

```python
uncertainty_node = 0.5

n_nodes = [1, 2, 4, 10, 100]
# n_nodes = 1

for n in n_nodes:
    unc_vector = np.ones(n) * uncertainty_node
    weights = np.ones(n) / n
#     var_vector = np.float_power(unc_vector, 2.0)
    total_var = np.dot(np.float_power(weights, 2.0), np.float_power(unc_vector, 2.0))
    print(n, total_var)
```
